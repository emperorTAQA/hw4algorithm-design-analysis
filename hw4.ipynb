## 问题 1 
**最小生成树（Minimum Spanning Tree）**

设  **G**  为一个带权重的连通无向图，且所有边的权重均不相等。令$e_i$ 为权重第 $i$ 小的边。最小生成树（MST）是否必须包含 $e_1$ ? 同理，是否必须包含 $e_2$ 和 $e_3$ ? 若必须包含，请给出证明；否则，请构造反例。需从基本原理论证，不能依赖割引理(cut lemma) 或 Prim/Kruskal算法的正确性。



### 分析 $(e_1)$

**结论：$(e_1)$ 必须包含在MST中。**

**证明：**
假设 $(G)$ 是一个带权重的连通无向图，且所有边的权重均不相等。设 $(e_1)$ 为权重最小的边。

1. **反证法**：假设 $(e_1)$ 不在MST中。
2. **构造新树**：如果 $(e_1)$ 不在MST中，那么我们可以将 $(e_1)$ 添加到MST中，形成一个环。
3. **删除最大权重边**：在这个环中，一定存在一条边 $(e')$，其权重大于 $(e_1)$ 的权重（因为 $(e_1)$ 是权重最小的边）。我们可以删除这条边 $(e')$，从而得到一个新的生成树。
4. **新树更优**：新的生成树的总权重比原来的MST小，这与MST是最小生成树的定义矛盾。

因此，$(e_1)$ 必须包含在MST中。

### 分析 $(e_2)$

**结论：$(e_2)$ 不一定必须包含在MST中。**

**反例：**
考虑以下简单图：

```

```

- 边 $(AC)$ 权重为 1。
- 边 $(AB)$ 权重为 2。
- 边 $(BD)$ 权重为 3。
- 边 $(CD)$ 权重为 4。

在这个图中：
- $(e_1 = AC)$，权重为 1。
- $(e_2 = AB)$，权重为 2。

MST 可以是 $(\{AC, AB, BD\})$ 或 $(\{AC, CD, BD\})$。在第二种情况下，$(e_2 = AB)$ 并不在MST中。

### 分析 $(e_3)$

**结论：$(e_3)$ 不一定必须包含在MST中。**

## 问题 2 
**瓶颈生成树（Bottleneck Spanning Tree）**

带有权重的无向图 $G(V,E,w)$ 的瓶颈生成树，表现为：在所有生成树中，最大权重边的权重值最小。即，BST $T$ 最小化瓶颈损失 $c(T)=max_{e \in T}{w(e)}$。

1. 证明 $G$ 的每一个最小生成树（MST）都是瓶颈生成树（BST）
2. 设计一个线性时间复杂度的算法：， 对于一个图 $G(V,E,w)$ 和一个整数 $b$，判断图 $ G$ 是否存在一个瓶颈生成树，其最大权重边的权重不超过 $b$，分析算法设计思路，并基于python编程实现。
3. 设计一个线性时间复杂度的算法：对于给定的图 $G(V,E,w)$，找到其瓶颈生成树，分析算法设计思路，并基于python编程实现。

idea：
### 1. 证明每一个最小生成树（MST）都是瓶颈生成树（BST）
设 $(T)$ 是图 $(G=(V, E, w))$ 的一个最小生成树，其最大权重边为 $(e_{max})$，权重为 $(w(e_{max}))$ 。
假设存在另一个生成树 $(T')$ ，其最大权重边的权重为 $(w'(e_{max}'))$ ，且 $(w'(e_{max}') < w(e_{max}))$ 。
考虑 $(T)$ 中去掉 $(e_{max})$ 后，图被分成两个连通分量 $(V_1)$ 和 $(V_2)$ 。由于 $(T')$ 是生成树，所以 $(T')$ 中必然存在一条边 $(e')$ 连接 $(V_1)$ 和 $(V_2)$ ，且 $(w(e') \leq w'(e_{max}') < w(e_{max}))$ 。
将 $(e')$ 加入到 $(T - \{e_{max}\})$ 中，会得到一个新的生成树 $(T'')$ ，且 $(T'')$ 的总权重小于 $(T)$ 的总权重，这与 $(T)$ 是最小生成树矛盾。所以，在所有生成树中，最小生成树的最大权重边的权重是最小的，即每一个最小生成树都是瓶颈生成树。


### 2. 判断图 \(G\) 是否存在一个瓶颈生成树，其最大权重边的权重不超过 \(b\)
#### 算法设计思路
采用深度优先搜索（DFS）或广度优先搜索（BFS）来遍历图，同时记录遍历过程中经过边的最大权重。从图中的任意一个顶点开始遍历，在遍历过程中，若遇到一条边的权重超过 \(b\) ，则停止对这条边所在分支的继续探索。如果能够遍历完所有顶点，说明存在一个瓶颈生成树，其最大权重边的权重不超过 \(b\) 。

#### 时间复杂度分析
假设图中有 \(n\) 个顶点和 \(m\) 条边，无论是 DFS 还是 BFS ，都需要遍历每个顶点和边，时间复杂度为 \(O(n + m)\) ，满足线性时间复杂度要求。

````python

from collections import deque

def has_bottleneck_spanning_tree(graph, b):
    visited = set()
    def bfs(start):
        queue = deque([start])
        visited.add(start)
        while queue:
            vertex = queue.popleft()
            for neighbor, weight in graph[vertex].items():
                if neighbor not in visited:
                    if weight > b:
                        continue
                    queue.append(neighbor)
                    visited.add(neighbor)
    start_vertex = list(graph.keys())[0]
    bfs(start_vertex)
    return len(visited) == len(graph)

# 测试图的表示，邻接表形式
graph = {
    'A': {'B': 2, 'C': 4},
    'B': {'A': 2, 'C': 1},
    'C': {'A': 4, 'B': 1}
}
b = 3
print(has_bottleneck_spanning_tree(graph, b))
````

### 3. 找到图 \(G\) 的瓶颈生成树
#### 算法设计思路
首先对图的边按照权重从小到大进行排序。然后从权重最小的边开始，依次尝试加入边到当前的生成森林中，同时使用并查集数据结构来维护连通性，避免形成环。在加入边的过程中，记录当前生成树中最大权重边的权重。当生成树包含所有顶点时，得到的就是瓶颈生成树。

#### 时间复杂度分析
对边排序的时间复杂度为 $(O(m \log m))$ ，其中 $(m)$ 是边的数量。之后遍历边的过程中，每次操作并查集的时间复杂度近似为常数，总共需要遍历 $(m)$ 条边，所以这部分时间复杂度为 $(O(m))$ 。整体时间复杂度为 $(O(m \log m))$ ，在边数 $(m)$ 与顶点数 $(n)$ 满足线性关系（即 $(m = O(n))$ ）时，也可视为线性时间复杂度。

````python
class UnionFind:
    def __init__(self, n):
        self.parent = list(range(n))
        self.rank = [0] * n

    def find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])
        return self.parent[x]

    def union(self, x, y):
        root_x = self.find(x)
        root_y = self.find(y)
        if root_x != root_y:
            if self.rank[root_x] > self.rank[root_y]:
                self.parent[root_y] = root_x
            elif self.rank[root_x] < self.rank[root_y]:
                self.parent[root_x] = root_y
            else:
                self.parent[root_y] = root_x
                self.rank[root_x] += 1

def bottleneck_spanning_tree(graph):
    # 把顶点字符串映射为整数
    vertex_mapping = {vertex: index for index, vertex in enumerate(graph.keys())}
    edges = []
    for vertex in graph:
        for neighbor, weight in graph[vertex].items():
            edges.append((weight, vertex_mapping[vertex], vertex_mapping[neighbor]))
    edges.sort()
    uf = UnionFind(len(graph))
    mst_edges = []
    max_weight = 0
    # 反向映射，用于将整数还原为顶点字符串
    reverse_mapping = {index: vertex for vertex, index in vertex_mapping.items()}
    for weight, u, v in edges:
        if uf.find(u) != uf.find(v):
            uf.union(u, v)
            # 将整数还原为顶点字符串
            mst_edges.append((reverse_mapping[u], reverse_mapping[v]))
            max_weight = weight
            if len(mst_edges) == len(graph) - 1:
                break
    return mst_edges, max_weight

# 测试图的表示，邻接表形式
graph = {
    'A': {'B': 2, 'C': 4},
    'B': {'A': 2, 'C': 1},
    'C': {'A': 4, 'B': 1}
}
mst_edges, max_weight = bottleneck_spanning_tree(graph)
print("瓶颈生成树的边:", mst_edges)
print("瓶颈生成树最大权重边的权重:", max_weight)
    
````

## 问题 3

**道路网（Road Network）**

假设有一个以图 $ G(V, E, l) $ 表示的道路网络，连接了一组城市 $ V $。我们假设该网络是有向的，并且每条道路 $(u, v) \in E$ 都有一个非负的长度 $ l(u, v) $。一条新的道路即将被建造，因此有一个列表 $ E' $ 包含它可以连接的城市对。每对 $(u, v) \in E'$ 都有一个对应的长度 $ l'(u, v) $。我们希望选择一对城市，使得两个城市 $ s, t \in V $ 之间的距离减少最大。请为此问题编写一个高效的算法，并详细解释算法的正确性和复杂度。


### 算法设计思路（参考大模型）
1. **计算初始距离矩阵**：使用弗洛伊德 - 沃肖尔（Floyd - Warshall）算法或迪杰斯特拉（Dijkstra）算法来计算原始道路网络中任意两个城市 \(s\) 和 \(t\) 之间的最短距离，得到距离矩阵 $(\text{dist})$ 。对于有向图，如果图中不存在负权边，Dijkstra 算法时间复杂度为 $(O(|V|^2 + |E|))$（邻接矩阵存储 ）或 $(O(|V|\log|V| + |E|))$（邻接表存储 ），Floyd - Warshall 算法时间复杂度为 \(O(|V|^3)\) ，这里我们优先考虑使用 Dijkstra 算法计算单源最短路径，循环 \(|V|\) 次来得到任意两点间最短路径。
2. **遍历新道路列表**：对于新道路列表 \(E'\) 中的每一对城市 \((u, v)\) ：
    - 尝试将这条新道路加入到图中，即更新图的边集。
    - 再次使用 Dijkstra 算法计算加入新道路后所有城市对之间的最短距离，得到新的距离矩阵 $(\text{new\_dist})$ 。
    - 计算每个城市对在加入新道路前后的距离减少量 $ \text{reduction} = \text{dist}[s][t] - \text{new\_dist}[s][t] $  。
3. **找出最大距离减少量**：在遍历完 \(E'\) 中所有城市对后，记录下使得距离减少量最大的那一对城市 \((u, v)\) 。

### 算法正确性证明
我们的目标是找到一对城市，使得加入连接它们的新道路后，某两个城市 \(s\) 和 \(t\) 之间的距离减少最大。通过计算原始距离矩阵，我们得到了没有新道路时城市间的最短距离。当遍历新道路列表时，每次加入一条新道路并重新计算距离矩阵，这样就能准确得到加入该新道路后城市间距离的变化情况。通过比较所有新道路加入后城市间距离的减少量，必然能找到使得距离减少最大的那一对城市，所以该算法是正确的。

### 算法复杂度分析
1. **计算初始距离矩阵**：若使用 Dijkstra 算法，假设图用邻接表存储，循环 \(|V|\) 次计算单源最短路径，每次时间复杂度为 $(O(|V|\log|V| + |E|))$ ，所以这部分时间复杂度为 $(O(|V|(|V|\log|V| + |E|)))$ 。
2. **遍历新道路列表**：对于新道路列表 \(E'\) 中的每一对城市 \((u, v)\) ，加入新道路后重新计算距离矩阵，每次也使用 Dijkstra 算法，时间复杂度为 $(O(|V|(|V|\log|V| + |E|)))$ ，而 \(E'\) 中最多有 \(|V|^2\) 对城市，所以这部分时间复杂度为 $O(|E'|\times|V|(|V|\log|V| + |E|))$ 。
3. **总的时间复杂度**：算法总的时间复杂度为 $(O(|V|(|V|\log|V| + |E|)+|E'|\times|V|(|V|\log|V| + |E|)) = O((1 + |E'|)|V|(|V|\log|V| + |E|)))$ 。 

````python
import heapq
import math


# Dijkstra算法计算单源最短路径
def dijkstra(graph, start):
    distances = {vertex: math.inf for vertex in graph}
    distances[start] = 0
    pq = [(0, start)]
    while pq:
        current_distance, current_vertex = heapq.heappop(pq)
        if current_distance > distances[current_vertex]:
            continue
        for neighbor, weight in graph[current_vertex].items():
            distance = current_distance + weight
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(pq, (distance, neighbor))
    return distances


# 计算原始图中所有城市对的最短距离
def compute_distances(graph):
    num_vertices = len(graph)
    dist = [[math.inf] * num_vertices for _ in range(num_vertices)]
    for vertex in graph:
        distances = dijkstra(graph, vertex)
        for neighbor in graph:
            dist[list(graph.keys()).index(vertex)][list(graph.keys()).index(neighbor)] = distances[neighbor]
    return dist


# 主算法函数
def find_best_new_road(graph, new_roads):
    num_vertices = len(graph)
    dist = compute_distances(graph)
    max_reduction = -math.inf
    best_pair = None
    for u, v, l_uv in new_roads:
        # 复制原始图并加入新道路
        new_graph = {k: v.copy() for k, v in graph.items()}
        new_graph[u][v] = l_uv
        new_dist = compute_distances(new_graph)
        for s in range(num_vertices):
            for t in range(num_vertices):
                reduction = dist[s][t] - new_dist[s][t]
                if reduction > max_reduction:
                    max_reduction = reduction
                    best_pair = (u, v)
    return best_pair


# 示例图，邻接表形式，键为顶点，值为邻接顶点及其边的长度
graph = {
    'A': {'B': 2, 'C': 4},
    'B': {'C': 1},
    'C': {}
}
# 新道路列表，元素为 (u, v, l'(u, v))
new_roads = [('A', 'C', 3)]
result = find_best_new_road(graph, new_roads)
print("使得距离减少最大的城市对:", result)
````

**逃离问题**

idea：
## 解题思路 参考大模型
### (1) 证明
**转换方法**：
对于给定的具有顶点和边容量的网络 $(G=(V, E))$ ，我们按如下方式构造新网络 $(G'=(V', E'))$ ：
- **顶点转换**：将 $(G)$ 中的每个顶点 $(v\in V)$ 拆分成两个顶点 $(v_{in})$ 和 $(v_{out})$ 。其中 $(v_{in})$ 用于接收流入 $(v)$ 的流量，$(v_{out})$ 用于发出从 $(v)$ 流出的流量。
- **边的调整**：
    - 对于 $(G)$ 中每一条边 $((u, v)\in E)$ ，在 \(G'\) 中创建边 $((u_{out}, v_{in}))$ ，其容量与 $(G)$ 中边 $((u, v))$ 的容量相同。
    - 在 $(v_{in})$ 和 $(v_{out})$ 之间添加一条边 $((v_{in}, v_{out}))$ ，其容量为顶点 $(v)$ 在 $(G)$ 中的容量。
 
**正确性证明**：
- **流量一致性**：在原网络 $(G)$ 中，顶点 $(v)$ 的流入流量受到其自身容量限制，同时边也有容量限制。在新网络 $(G')$ 中，流入 $(v_{in})$ 的流量必须通过边 $((v_{in}, v_{out}))$ 才能从 $(v_{out})$ 流出，而边 $((v_{in}, v_{out}))$ 的容量恰好是顶点 $(v)$ 在原网络中的容量，这就保证了顶点容量限制的等价性。对于边的容量，由于只是改变了连接方式，边的容量数值并未改变，所以边容量限制也保持一致。因此，两个网络上的最大流必然相同。
- **规模分析**：
    - **顶点数量**：新网络 $(G')$ 中，因为每个原顶点 $(v)$ 被拆分成 $(v_{in})$ 和 $(v_{out})$ ，所以 $(|V'| = 2|V|)$ ，满足 $(V' = O(V))$ 。
    - **边的数量**：原网络 $(G)$ 中的每条边 $((u, v))$ 对应新网络 $(G')$ 中的边 $((u_{out}, v_{in}))$ ，并且每个顶点拆分后增加了一条边 $((v_{in}, v_{out}))$ ，所以 $(|E'| = |E| + |V|)$ 。在连通图中，$(|V| = O(|E|))$ ，所以 $(|E'| = O(|E|))$ ，即 $(E' = O(E))$ 。

### (2) 算法及运行时间分析
**算法描述**：
1. **网络构建**：
    - 按照上述 (1) 中的方法，将具有顶点和边容量的 $(n\times n)$ 网格图（即网络 $(G)$ ）转换为仅具有边容量的网络 $(G')$ 。
    - 新增源点 $(s)$ ，从源点 $(s)$ 向每个起点 $((x_i, y_i))$ 对应的 $(v_{out})$ 顶点（在转换后的网络 $(G')$ 中）连接一条容量为 $(1)$ 的边。这是因为每个起点只能作为一条不相交路径的起始点，流量为 $(1)$ 。
    - 新增汇点 $(t)$ ，将边界上的每个顶点对应的 $(v_{in})$ 顶点（在转换后的网络 $(G')$ 中）向汇点 $(t)$ 连接一条容量为 $(1)$ 的边。这是为了接收从起点出发并通过不相交路径到达边界的流量。
2. **最大流计算**：使用高效的最大流算法，如Dinic算法来计算从源点 $(s)$ 到汇点 $(t)$ 的最大流。Dinic算法通过分层图和阻塞流的概念，能够在 $(O(V^2E))$ 的时间复杂度内计算出最大流。在我们的 $(n\times n)$ 网格图中，$(|V| = n^2)$ ，$(|E| = O(n^2))$ 。
3. **结果判断**：如果计算得到的最大流值等于 $(m)$ ，则说明存在 $(m)$ 条顶点不相交的路径，即存在逃离方案；否则不存在。

**运行时间分析**：
- **网络构建时间**：将顶点拆分、添加源点和汇点以及相关边的操作，时间复杂度为 $(O(V + E))$ 。在 $(n\times n)$ 网格图中，$(V = n^2)$ ，$(E = O(n^2))$ ，所以这部分时间复杂度为 $(O(n^2))$ 。
- **最大流计算时间**：使用Dinic算法计算最大流，其时间复杂度为 $(O(V^2E))$ 。代入 $(V = n^2)$ ，$(E = O(n^2))$ ，可得这部分时间复杂度为 $(O(n^6))$ 。
- **总的时间复杂度**：算法的总时间复杂度由最大流计算部分主导，为 $(O(n^6))$ 。 

````python
import collections


# 构建分层图
def bfs(net, source, sink, level):
    for vertex in level:
        level[vertex] = -1
    level[source] = 0
    queue = collections.deque([source])
    while queue:
        u = queue.popleft()
        for v in net[u]:
            if level[v] < 0 and net[u][v] > 0:
                level[v] = level[u] + 1
                queue.append(v)
    return level[sink] >= 0


# 寻找增广路径并更新流量
def dfs(net, u, sink, flow, level):
    if u == sink:
        return flow
    for v in net[u]:
        if level[v] == level[u] + 1 and net[u][v] > 0:
            min_flow = dfs(net, v, sink, min(flow, net[u][v]), level)
            if min_flow > 0:
                net[u][v] -= min_flow
                # 若反向边不存在则初始化
                if u not in net[v]:
                    net[v][u] = 0
                net[v][u] += min_flow
                return min_flow
    return 0


# Dinic算法计算最大流
def dinic(net, source, sink):
    max_flow = 0
    level = {vertex: -1 for vertex in net}
    while bfs(net, source, sink, level):
        flow = dfs(net, source, sink, float('inf'), level)
        while flow > 0:
            max_flow += flow
            flow = dfs(net, source, sink, float('inf'), level)
    return max_flow


# 构建逃离问题的网络
def build_network(n, start_points):
    # 顶点拆分后的编号映射，(i, j) -> (in编号, out编号)
    vertex_mapping = {}
    index = 0
    for i in range(1, n + 1):
        for j in range(1, n + 1):
            vertex_mapping[(i, j)] = (index, index + 1)
            index += 2

    network = collections.defaultdict(dict)
    # 内部顶点连接
    for i in range(1, n + 1):
        for j in range(1, n + 1):
            u_in, u_out = vertex_mapping[(i, j)]
            # 顶点自身的in和out连接，容量为1（表示顶点不相交）
            network[u_in][u_out] = 1
            # 初始化反向边
            network[u_out][u_in] = 0
            if i > 1:
                v_in, v_out = vertex_mapping[(i - 1, j)]
                network[u_out][v_in] = 1
                network[v_in][u_out] = 0
            if i < n:
                v_in, v_out = vertex_mapping[(i + 1, j)]
                network[u_out][v_in] = 1
                network[v_in][u_out] = 0
            if j > 1:
                v_in, v_out = vertex_mapping[(i, j - 1)]
                network[u_out][v_in] = 1
                network[v_in][u_out] = 0
            if j < n:
                v_in, v_out = vertex_mapping[(i, j + 1)]
                network[u_out][v_in] = 1
                network[v_in][u_out] = 0

    source = index
    sink = index + 1
    network[source] = {}
    network[sink] = {}
    # 连接源点到起点
    for start in start_points:
        x, y = start
        u_in, u_out = vertex_mapping[(x, y)]
        network[source][u_out] = 1
        network[u_out][source] = 0
    # 连接边界顶点到汇点
    for i in range(1, n + 1):
        for j in range(1, n + 1):
            if i == 1 or i == n or j == 1 or j == n:
                u_in, u_out = vertex_mapping[(i, j)]
                network[u_in][sink] = 1
                network[sink][u_in] = 0
    return network, source, sink


# 解决逃离问题
def escape_problem(n, start_points):
    network, source, sink = build_network(n, start_points)
    max_flow = dinic(network, source, sink)
    return max_flow == len(start_points)


# 测试示例
n = 4
start_points = [(2, 2), (3, 3)]
print(escape_problem(n, start_points))
    
````
